```python
import pandas as pd
from scipy.stats import chi2_contingency
import pandas as pd

# Replace 'your_dataset.xlsx' with the actual file path to your Excel file
excel_file_path = 'v2_all ncRNA-disease information.xlsx'

# Load the dataset from Excel into a Pandas DataFrame
df = pd.read_excel(excel_file_path)

# Display the DataFrame
df.head()






# Function to calculate Cramér's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

# Calculate Cramér's V for all pairs of columns
columns = df.columns
cramers_matrix = pd.DataFrame(index=columns, columns=columns, dtype=float)
for i in range(len(columns)):
    for j in range(len(columns)):
        cramers_matrix.iloc[i, j] = cramers_v(df[columns[i]], df[columns[j]])

# Print the Cramér's V matrix
print(cramers_matrix)
```
