```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Convert images to grayscale
x_train_gray = np.dot(x_train[..., :3], [0.2989, 0.5870, 0.1140])
x_test_gray = np.dot(x_test[..., :3], [0.2989, 0.5870, 0.1140])

# Reshape the data
x_train_flat = x_train_gray.reshape(x_train_gray.shape[0], -1)
x_test_flat = x_test_gray.reshape(x_test_gray.shape[0], -1)

# Standardize the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_flat)
x_test_scaled = scaler.transform(x_test_flat)

# Define Random Forest classifier
rf_classifier = RandomForestClassifier()

# Define hyperparameters for Grid Search
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20]
}

# Perform Grid Search
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5)
grid_search.fit(x_train_scaled, y_train.ravel())

# Get the best parameters
best_params = grid_search.best_params_

# Train the model with best parameters
best_rf_classifier = RandomForestClassifier(n_estimators=best_params['n_estimators'], max_depth=best_params['max_depth'])
best_rf_classifier.fit(x_train_scaled, y_train.ravel())

# Evaluate on validation set
y_pred_val = best_rf_classifier.predict(x_test_scaled)
val_accuracy = accuracy_score(y_test.ravel(), y_pred_val)
val_precision = precision_score(y_test.ravel(), y_pred_val, average='macro')
val_recall = recall_score(y_test.ravel(), y_pred_val, average='macro')

print("Validation Accuracy:", val_accuracy)
print("Validation Precision:", val_precision)
print("Validation Recall:", val_recall)

# Evaluate on test set
y_pred_test = best_rf_classifier.predict(x_test_scaled)
test_accuracy = accuracy_score(y_test.ravel(), y_pred_test)
test_precision = precision_score(y_test.ravel(), y_pred_test, average='macro')
test_recall = recall_score(y_test.ravel(), y_pred_test, average='macro')

print("\nTest Accuracy:", test_accuracy)
print("Test Precision:", test_precision)
print("Test Recall:", test_recall)
```

```python
import tensorflow_datasets as tfds
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Load IMDb dataset from TensorFlow Datasets
imdb, info = tfds.load('imdb_reviews', with_info=True, as_supervised=True)

# Split the data into train and test sets
train_data, test_data = imdb['train'], imdb['test']

# Initialize empty lists to store sentences and labels
train_sentences = []
train_labels = []
test_sentences = []
test_labels = []

# Extract sentences and labels from the TensorFlow Datasets
for s, l in train_data:
    train_sentences.append(str(s.numpy()))
    train_labels.append(l.numpy())

for s, l in test_data:
    test_sentences.append(str(s.numpy()))
    test_labels.append(l.numpy())

# Convert labels to numpy arrays
train_labels = np.array(train_labels)
test_labels = np.array(test_labels)

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000)

# Transform text data to TF-IDF features
train_features = tfidf_vectorizer.fit_transform(train_sentences)
test_features = tfidf_vectorizer.transform(test_sentences)

# Initialize KNN classifier
knn_model = KNeighborsClassifier(n_neighbors=5)

# Train KNN model
knn_model.fit(train_features, train_labels)

# Evaluate KNN model
accuracy_knn = knn_model.score(test_features, test_labels)
print("KNN Model Accuracy:", accuracy_knn)

# Initialize Decision Tree classifier
dt_model = DecisionTreeClassifier()

# Define hyperparameters to tune
param_grid = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Perform Grid Search with Cross-Validation
grid_search = GridSearchCV(estimator=dt_model, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(train_features, train_labels)

# Get the best model
best_dt_model = grid_search.best_estimator_

# Evaluate the best model
accuracy_dt = best_dt_model.score(test_features, test_labels)
print("Decision Tree Model Accuracy:", accuracy_dt)
```
