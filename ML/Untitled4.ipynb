```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('apple_quality.csv')
```

```python
print(df.head())
```

```python
correlation_matrix = df.corr()
```

```python
correlation_matrix = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Apple Quality Dataset')
plt.show()
```

```python
column1_name = 'Crunchiness'  # Replace 'column1' with the actual name of the first column
column2_name = 'Juiciness'# Replace 'column2' with the actual name of the second column

# Clip values of the selected columns to be within the range (-1, 1)
df[column1_name] = df[column1_name].clip(lower=-1, upper=1)
df[column2_name] = df[column2_name].clip(lower=-1, upper=1)

# Create a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x=column1_name, y=column2_name)
plt.title(f'Scatter Plot of {column1_name} vs {column2_name} (Clipped)')
plt.xlabel(column1_name + ' (Clipped)')
plt.ylabel(column2_name + ' (Clipped)')
plt.grid(True)
plt.show()
```
